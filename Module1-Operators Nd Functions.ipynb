# 1.10 Strings in Python

## 1.10.1 Immutable Nature of Strings
- Strings are **immutable** in Python.
- Once a string is created, its **characters cannot be changed individually**.

```python
str1 = "Hello"
# str1[0] = "h"  âŒ This will raise an error
```

        - `\'` â†’ Single quote

# 1.10.3 String Operations

## 1.10.3.1 Concatenation
- Join two or more strings using the `+` operator:

```python
str1 = "Hello"
str2 = "World"
result = str1 + " " + str2
print(result)  # Output: Hello World
```

 ## 1.10.3.2 Length of a String
 - Use the len() function to find the number of characters in a string:

```python
 str1 = "Hello World"
print(len(str1))  # Output: 11
```

 ## 1.10.3.3 Indexing
 - Access individual characters using 0-based indexing:

```python
  str1 = "Hello World"
print(str1[0])  # 'H'
print(str1[1])  # 'e'
print(str1[-1]) # 'd' (last character)
```

 ## 1.10.3.4 Slicing
 - Access a substring using [start: end] notation:
    - start â†’ starting index (inclusive)
    - end â†’ ending index (exclusive)

```python
str1 = "Hello World"
print(str1[1:4])  # 'ell'
print(str1[0:4])  # 'Hell'
print(str1[1:len(str1)])  # 'ello World'
```

## 1.10.4 Common String Methods

| Method | Description | Example | Result |
|---|---|---|---|
| `str.endswith(sub)` | Returns `True` if the string ends with `sub` | `"Hello".endswith("lo")` | `True` |
| `str.capitalize()` | Capitalizes the first character of the string | `"hello".capitalize()` | `'Hello'` |
| `str.replace(old, new)` | Replaces all occurrences of `old` with `new` | `"hello world".replace("world", "Python")` | `'hello Python'` |
| `str.find(sub)` | Returns the **index of the first occurrence** of `sub` | `"hello".find("l")` | `2` |
| `str.count(sub)` | Counts the **occurrences** of `sub` in the string | `"banana".count("a")` | `3` |

 # 1.12 Lists in Python

## 1. Introduction
- Lists are a **built-in data type** that stores a **collection of values**.
- Can store **elements of different types** (int, float, string, etc.).
- Lists are **ordered** and **mutable** (can be changed after creation).

## 2. Creating Lists

```python
# List of marks (integers)
marks = [87, 64, 33, 95, 76]

# List of students (strings)
students = ["Kazan", 85, "Delhi"]

# Accessing elements
print(marks[0])      # Output: 87
print(students[0])   # Output: 'Kazan'

# Length of a list
print(len(marks))    # Output: 5

#slicing
print(marks[1:4])
```

## 3. Common List Methods

| Method | Description | Example | Result |
|---|---|---|---|
| `list.append(ele)` | Adds an element at the end | `marks.append(100)` | `[87, 64, 33, 95, 76, 100]` |
| `list.sort()` | Sorts list in ascending order | `marks.sort()` | `[33, 64, 76, 87, 95]` |
| `list.sort(reverse=True)` | Sorts list in descending order | `marks.sort(reverse=True)` | `[95, 87, 76, 64, 33]` |
| `list.reverse()` | Reverses the order of elements | `marks.reverse()` | `[76, 95, 33, 64, 87]` |
| `list.insert(idx, ele)` | Inserts element at a specific index | `marks.insert(2, 50)` | `[87, 64, 50, 33, 95, 76]` |
| `list.remove(ele)` | Removes first occurrence of element | `marks.remove(64)` | `[87, 33, 95, 76]` |
| `list.pop(idx)` | Removes element at index (default last) | `marks.pop(1)` | `[87, 33, 95, 76]` |

 # 1.13 Tuples
## 1. Introduction
- Tuples are a **built-in data type** that stores a **collection of values**.
- Tuples are **ordered** and **immutable** (cannot be changed after creation).
- Can store **elements of different types** (int, float, string, etc.).

```python
# Tuple of marks (integers)
marks = (87, 64, 33, 95, 76)

# Tuple of students (strings)
students = ("Kazan", "Amit", "Delhi")

# Accessing elements
print(marks[0])      # Output: 87
print(students[0])   # Output: 'Kazan'

# Length of a tuple
print(len(marks))    # Output: 5

# Slicing a tuple
print(marks[1:4])    # Output: (64, 33, 95)
```

 ## 2. Tuple Characteristics
 - Immutable: You cannot add, remove, or modify elements.
 - Supports indexing and slicing.
 - Can contain duplicate elements.
 - Can be nested (tuple inside a tuple).

```python
nested_tuple = (1, 2, (3, 4), 5)
print(nested_tuple[2])      # Output: (3, 4)
print(nested_tuple[2][1])   # Output: 4
```

## 3. Common Tuple Methods & Operations

| Method / Function | Description | Example | Result |
|---|---|---|---|
| `tuple.count(ele)` | Returns the number of occurrences of an element | `(1, 2, 3, 2).count(2)` | `2` |
| `tuple.index(ele)` | Returns the index of the first occurrence of an element | `(1, 2, 3, 2).index(2)` | `1` |
| `len(tuple)` | Returns the length of the tuple | `len((1, 2, 3))` | `3` |

**Note:** Tuples do not have methods like `append()`, `remove()`, or `sort()` because they are immutable.

 ## 4. Converting Between Lists and Tuples

```python
# List to tuple
my_list = [1, 2, 3]
my_tuple = tuple(my_list)
print(my_tuple)  # Output: (1, 2, 3)

# Tuple to list
new_list = list(my_tuple)
print(new_list)  # Output: [1, 2, 3]
```

# 1.14. Looping in Python
Loops are used to **execute a block of code repeatedly** until a certain condition is met.

## 1. `while` Loop
- Repeats a block of code **as long as a condition is True**.

```python
count = 0
while count < 5:
    print("Count:", count)
    count += 1
```

 ## 2.`for` Loop
 - Iterates over a sequence (list, tuple, string, or range).

```python
# Iterating over a list
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)

# Iterating using range()
for i in range(5):  # 0 to 4
    print(i)
```

 ## 3. `range()` Function
 - Generates a sequence of numbers.
 - Syntax: range(start, stop, step)
 - start â†’ starting number (default 0)
 - stop â†’ stop before this number
 - step â†’ increment (default 1)

```python
for i in range(1, 10, 2):
    print(i)
```

 ## 4. `break` Statement
 - Exits the current loop immediately, even if the condition is True.

```python
for i in range(5):
    if i == 3:
        break
    print(i)
```

 ## 5. `continue` Statement
 - Skips the current iteration and moves to the next iteration.

```python
for i in range(5):
    if i == 2:
        continue
    print(i)
```

 ## 6. Nested Loops
 - A loop inside another loop.

```python
for i in range(1, 4):
    for j in range(1, 3):
        print("i =", i, "j =", j)
```

# 1.15. Python Functions
Functions are **reusable blocks of code** that perform a specific task. They help in **modularity** and **code reusability**.

## 1. Defining a Function

```python
def greet():
    print("Hello, welcome to Python!")
```

 - `def` â†’ keyword to define a function.
 - `greet` â†’ function name.
 - `()` â†’ parentheses (can include parameters).
 - The code inside the function is indented.

 ## 2. Calling a Function

```python
greet()  # Output: Hello, welcome to Python!
```

## 3. Function with Parameters
Parameters allow passing information into functions.

```python
def greet_person(name):
    print("Hello", name)

greet_person("Amit")  # Output: Hello Amit
greet_person("John")  # Output: Hello John
```

 ## 4. Function with Return Value
`return` sends a value back to the caller.

```python
def add(a, b):
    return a + b

result = add(5, 3)
print(result)  # Output: 8
```

 ## 5. Default Parameters
Default parameters are used if no argument is provided.

```python
def greet(name="Guest"):
    print("Hello", name)

greet()        # Output: Hello Guest
greet("Amit")  # Output: Hello Amit
```

 ## 6. Keyword Arguments
Arguments can be passed by keyword, ignoring the order.

```python
def describe_person(name, age):
    print("Name:", name)
    print("Age:", age)

describe_person(age=25, name="John")
```

 ## 7. Variable-Length Arguments

 ### a) Arbitrary Positional Arguments `(*args)`

```python
def add_numbers(*args):
    total = 0
    for num in args:
        total += num
    return total

print(add_numbers(1, 2, 3, 4))  # Output: 10
```

 ### b) Arbitrary Keyword Arguments `(**kwargs)`

```python
def student_info(**kwargs):
    for key, value in kwargs.items():
        print(key, ":", value)

student_info(name="Amit", age=20, city="Delhi")
```

 ## 8. Lambda Functions (Anonymous Functions)
Lambda functions are one-line functions without a name.

```python
square = lambda x: x**2
print(square(5))  # Output: 25

add = lambda a, b: a + b
print(add(2, 3))  # Output: 5
```

 ## 9. Scope of Variables
 - Local variables exist only inside the function.
 - Global variables can be accessed anywhere.

```python
def func():
    local_var = 10   # Local variable
    print(local_var)

global_var = 5       # Global variable
func()
print(global_var)    # Output: 5
```
